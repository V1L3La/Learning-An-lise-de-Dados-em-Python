# Carrega arquivo Resultados da eleição 2018

```python
# Vamos carregar o arquivo com o resultado da eleição no Brasil dividido por municipio/zona
df_votacao = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Mapa_Geopandas/votacao_candidato_munzona_2018_BR.csv",
                         sep=';', encoding='latin1')
df_votacao.head().T
```

```python
# separar dados de 2 turno (Somente com Jair Bonsonaro e Haddad)
df_votacao = df_votacao[(df_votacao['NR_TURNO'] == 2)]
df_votacao.sample(4).T
```

```python
# Verificar quantos municipios unicos exitem no dataframe pelo código do municipio
df_votacao['CD_MUNICIPIO'].value_counts()
```

```python
# Verificar quantos municipios unicos exitem no dataframe pelo código do municipio
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Quantas cidades fora do Brasil    
df_votacao[df_votacao["SG_UF"] == 'ZZ'] ['CD_MUNICIPIO'].nunique()
```

```python
# Eliminar os municipios de fora do brasil
df_votacao = df_votacao[df_votacao['SG_UF'] != 'ZZ']
df_votacao.head().T
```

```python
# Verificando a quantidade de municipios
df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Os dados estão separados por municipio/zona
# Municipios pequenos normalmente tem apenas 1 zona eleitoral
# mas municipios maiores podem ter várias zonas eleitorias

# Verificando Belo Horizonte
# Verificar quantas zonas eleitorais tem em Belo Horizonte     NR_ZONA
#df_votacao[df_votacao["NM_MUNICIPIO"] == 'BELO HORIZONTE']['NR_ZONA'].nunique()
df_votacao[df_votacao["NM_MUNICIPIO"] == 'BELO HORIZONTE']
```

```python
# Eliminar algumas colunas do dataframe que não serão utilizados

df_votacao = df_votacao[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS' ]]
df_votacao.head()
```

```python
df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO','NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum()
```

```python
# Precisamos agregar os dados por municipios e candidatos
df_votacao = df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_URNA_CANDIDATO']).sum().reset_index()
df_votacao.sample(6)
```

```python
# Verificando Belo Horizonte
# Fazer o groupby, reindexar o indice, vai salvar isso em um dataframe comum
df_votacao[df_votacao["NM_MUNICIPIO"] == 'BELO HORIZONTE']
```

## Organizando o vencedor em cada municipio

```python
# Ordenar o dataframe pela quantidade de votos e depois deixar apenas o primeiro
# registro de cada município
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False).head(10)
```

```python
# vamos ordenar
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)

# dropa as linhas duplicadas, mantendo a primeira
df_votacao.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)

df_votacao.head(10)
```

```python
# Verificando Belo Horizonte
# Fazer o groupby, reindexar o indice, vai salvar isso em um dataframe comum
df_votacao[df_votacao["NM_MUNICIPIO"] == 'BELO HORIZONTE']
```

```python
# criar uma coluna 'RESULTADO' representando o resultado da eleição naquele municipio
# sendo que o resultado ´True se o candidato Bolsonaro venceu naquele municipio
# e false em caso contrário
# Função lambda - retorna um x(True) quando linha(x) for = 17
df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x: x == 17)
```
