```python
#Valor mais alto em numa linha no dataframe
df[df['mensalidade'] == df['mensalidade'].max()]   
```

```python
#Valor mais baixo numa linha no dataframe
df[df['mensalidade'] == df['mensalidade'].min()]   
```

```python
#Valor mais baixo numa linha no dataframe com uma coluna
df[df['mensalidade'] == df['mensalidade'].min()] ['uf_busca']   
```

```python
#Valor mais baixo numa linha no dataframe com uma coluna
df[df['mensalidade'] == df['mensalidade'].min()] [['uf_busca', 'curso_busca', 'universidade_nome' ]]   
```
