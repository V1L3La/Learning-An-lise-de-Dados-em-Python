```python
import pandas as pd
# . é igual a "acessar funções da biblioteca...." 
df = pd.read_csv('https://raw.githubusercontent.com/V1L3La/Learning-An-lise-de-Dados-em-Python/main/Transa%25C3%25A7%25C3%25B5es_Imobili%25C3%25A1rias_-_ITBI.csv')  
df
```

```python
df.rename(columns={'NOME': 'NOME_DO_BAIRRO'}, inplace = True)   # troca o nome de uma coluna
df
```

```python
df['NOME_DO_BAIRRO'].unique()    # valores unicos(sem repetir) na coluna especificada
```

```python
  df['NOME_DO_BAIRRO'].value_counts()              # conta valores totais DENTRO da coluna  e  [] = selecionar
```

```python
Pesquisa_VIG = df.query('VALOR_RECOLHIDO < 3000.00 & NOME_DO_BAIRRO =="VIGARIO GERAL"').head()
Pesquisa_VIG
```

```python
Pesquisa_VIG.value_counts().plot.area()
```

```python
df
```

```python
df_Vigaris = df[df["NOME_DO_BAIRRO"] == 'VIGARIO GERAL']    # cria um novo dataframe a partir do antigo
df_Vigaris
```

```python
Pesquisa_VIG = df.query('NOME_DO_BAIRRO =="VIGARIO GERAL" & UTILIZACAO =="NAO RESIDENCIAL"').head()
Pesquisa_VIG
```

```python
df_Vigaris['TIPOLOGIA'].unique()
```

```python
# Criar subconjunto ou sub divisão

df_listagem_cordovil = df[df['NOME_DO_BAIRRO'] == 'CORDOVIL']
df_listagem_cordovil.head()
```
